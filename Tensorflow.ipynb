# Sessions:
Sessions are created and executed to do the actual computation. In tensorflow, creating constants/variables and applying operations doesn't actually compute those operations on them. We need to create and run sessions for them.

## Creating Sessions:
Sessions are created just by instantiating `Session` class.

```python
sess=tf.Session()
```

## Executing Sessions:
Sessions are executed using `.run()` method.

```python
sess.run(name)
```

## Terminating Sessions:
Sessions can be terminated using `.close()` method.

```python
sess.close()
```

__Note__: We can also use `with-as` python keyword to execute sessions. This avoids writing a terminating session statement.

# Tensors
A tensor is a generalization of vectors and matrices to potentially higher dimensions. Internally, TensorFlow represents tensors as n-dimensional arrays of base datatypes.

## Creating Tensors:

```python
sess=tf.Session()
```

### tf.fill()
``tf.fill()`` takes two arguments. One being the shape(tuple) of the tensor and other being the value. This functions creates a tensor of shape given, with all values equal to the value provided.

```python
sess.run(tf.fill((3,5),10))
```

### tf.zeros()
``tf.zeros()`` takes shape(tuple) as its argument. This functions creates a tensor of shape given, with all values equal to the 0.

```python
sess.run(tf.zeros((4,2)))
```

### tf.ones()
``tf.ones()`` takes shape(tuple) as its argument. This functions creates a tensor of shape given, with all values equal to the 1.

```python
sess.run(tf.ones((2,2)))
```

### tf.random_normal()
`tf.random_normal()` takes shape(tuple) as its argument. In addition, it also takes `mean` and `standard deviation`.  This functions creates a tensor of shape given, with random values normally distributed having given mean & standard deviation.

```python
sess.run(tf.random_normal((3,5),mean=6,stddev=3))
```

### tf.random_uniform()
`tf.random_uniform()` takes shape(tuple) as its argument. In addition, it also takes `minimum value` and `maximum value`.  This functions creates a tensor of shape given, with random values uniformly distributed from given minimum value to given maximum value.

```python
sess.run(tf.random_uniform((3,5),minval=-10,maxval=10))
```

### tf.random_poisson()
`tf.random_uniform()` takes shape(tuple) & `lambda` as its argument. This functions creates a tensor of shape given, with random values following poisson distribution with given lambda.

```python
sess.run(tf.random_poisson(3,(3,5)))
```

# Tensorflow Variables :
A TensorFlow variable is the best way to represent shared, persistent state manipulated by your program.

## Creating Variables:

### tf.Variable()
`tf.Variable()` takes a value as its argument and creates a variable with that value.

```python
var=tf.Variable([[9,8,7],[6,5,4]])
```

```python
sess.run(var)
```

### tf.global_variables_initializer()
__Note:__ Declared and assigned variables do not get executed using a session, until and unless they are initialized. To intialize, we use `tf.global_variables_initializer()` function.

```python
sess.run(tf.global_variables_initializer())
```

```python
sess.run(var)
```

### tf.get_variable()
`tf.get_variable()` function takes 3 arguments `name`, `shape` & `initializer`. This function creates a variable with given `shape` and value(`initializer`) provided. If `initializer` is not provided, then it initializes with random values.

```python
variable = tf.get_variable("variable",(1,2,3),initializer=tf.constant_initializer(7))
```

```python
sess.run(tf.global_variables_initializer())
sess.run(variable)
```

# Tensorflow Placeholders:
Placeholders are initially empty & are used to feed in the actual training data. However, they do need a declared expected data type with an optional shape argument.

## Creating placeholders:

### tf.placeholder()
`tf.placeholder()` takes `data type` & `shape` as its arguments & creates a placeholder of given data type & shape.

```python
placeholder=tf.placeholder(tf.float32,(None,6))
```

__Note:__ In above example, we have set `None` as we exactly don't know how many samples we are feeding in.

```python
sess.run(placeholder,feed_dict={placeholder:[[1,2,3,4,5,6]]})
```

# TensorFlow's Model Training & Testing Procedure:

## Data Loading & Cleaning:

Data:
https://archive.ics.uci.edu/ml/datasets/pima+indians+diabetes

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
diabetes = pd.read_csv('pima-indians-diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
cols_to_norm = ['Number_pregnant','Glucose_concentration',
                'Blood_pressure','Triceps','Insulin','BMI','Pedigree']
```

```python
diabetes[cols_to_norm]=diabetes[cols_to_norm].apply(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
diabetes.head()
```

## Feature Selection:

### Continuous Features

* Number of times pregnant
* Plasma glucose concentration a 2 hours in an oral glucose tolerance test
* Diastolic blood pressure (mm Hg)
* Triceps skin fold thickness (mm)
* 2-Hour serum insulin (mu U/ml)
* Body mass index (weight in kg/(height in m)^2)
* Diabetes pedigree function

### tf.feature_column.numeric_column()
`tf.feature_column.numeric_column()` takes in `key` and `shape` of the input. The given column must be a numeric column. This function considers the given column as a Numerical Feauture column. 

```python
num_preg=tf.feature_column.numeric_column('Number_pregnant')
plasma_gluc=tf.feature_column.numeric_column('Glucose_concentration')
dias_press=tf.feature_column.numeric_column('Blood_pressure')
tricep=tf.feature_column.numeric_column('Triceps')
insulin=tf.feature_column.numeric_column('Insulin')
bmi=tf.feature_column.numeric_column('BMI')
diabetes_pedigree=tf.feature_column.numeric_column('Pedigree')
age=tf.feature_column.numeric_column('Age')
```

### Categorical Features
If you know the set of all possible feature values of a column and there are only a few of them, you can use `categorical_column_with_vocabulary_list`. If you don't know the set of possible values in advance you can use `categorical_column_with_hash_bucket`

### tf.feature_column.categorical_column_with_vocabulary_list()
`tf.feature_column.categorical_column_with_vocabulary_list()` takes a `key` and a `vocabulary list` as its input arguments. `Vocabulary list` contains unique values that comes in the gven column. The given column must be a categorical column. This function considers the given column as a Categorical Feauture column. 

```python
assigned_group_1=tf.feature_column.categorical_column_with_vocabulary_list('Group',['A','B','C','D'])
```

### tf.feature_column.categorical_column_with_hash_bucket()
`tf.feature_column.categorical_column_with_hash_bucket()` takes a `key` & `hash bucket size` as its arguments. `Hash bucket size` is the number of unique values that the given colummn has. One can even provide a number which is greater than the number of categories. The given column must be a categorical column. This function considers the given column as a Categorical Feauture column.

```python
assigned_group_2=tf.feature_column.categorical_column_with_hash_bucket('Group', hash_bucket_size=10)
```

### tf.feature_column.embedding_column()
`tf.feature_column.embedding_column()` takes a `categorical column` & `dimension` as its arguments. This function is only useful when our model is dealing with the __neural networks__.

```python
embedded_group_column = tf.feature_column.embedding_column(assigned_group_1, dimension=4)
```

### Converting Continuous to Categorical

### tf.feature_column.bucketized_column()
`tf.feature_column.bucketized_column()` takes in `source column` & `boundaries` as its arguments. The function categorizes numerical data to categories.

```python
age_buckets = tf.feature_column.bucketized_column(age, boundaries=[20,30,40,50,60,70,80])
```

### Feature Columns:

```python
feat_cols=[num_preg,plasma_gluc,dias_press,tricep,insulin,bmi,diabetes_pedigree,assigned_group_1,age_buckets]
```

```python
x_data=diabetes.drop('Class',axis=1)
labels=diabetes['Class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_data,labels,test_size=0.33,random_state=42)
```

## Input Function:
An input function defines the input of your model. It contains type of input (Pandas or NumPy), input values, batch size, number of iterations,etc.

### tf.estimator.inputs.pandas_input_fn()
`tf.estimator.inputs.pandas_input_fn()` takes `feature matrix`, `vector of predictions`, `batch size`, `number of iterations`, `shuffle`,etc. The `feature matrix` & `vector of predictions` must be of pandas type. This lets the model to train/test in `batches` in given `iterations`. `Shuffle` lets to randomize inputs.

```python
input_train_func=tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
input_test_func=tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1000,shuffle=False)
```

### tf.estimator.inputs.numpy_input_fn()
This is same as `tf.estimator.inputs.pandas_input_fn()` but in place of pandas it takes numpy type of data.

## Model Creation:

### tf.estimator.LinearClassifier()
`tf.estimator.LinearClassifier()` takes in `featue columns` & `no of classes` as its arguments. This instantiates a Linear Classifier model for given `featue columns` & `no of classes`. This classifier works linearly.

```python
model=tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

### tf.estimator.DNNClassifier()
`tf.estimator.DNNClassifier()` is a Neural network based Classifier.

### tf.estimator.DNNLinearCombinedClassifier()
`tf.estimator.DNNLinearCombinedClassifier()` is a Classifier which is a combination of Linear Classifier & Neural network based Classifier.

### tf.estimator.LinearRegressor()
`tf.estimator.LinearRegressor()` is a Linear Regressor.

### tf.estimator.DNNRegressor()
`tf.estimator.DNNRegressor()` is a Neural network based Regressor.

### tf.estimator.DNNLinearCombinedRegressor()
`tf.estimator.DNNLinearCombinedRegressor()` is a Regressor which is a combination of Linear Regressor & Neural network based Regressor.

## Model Training:

### model.train()
`.train()` method takes an `input function` & `number of steps`  as its function parameters. This method fits the model for the given input data.

```python
model.train(input_fn=input_test_func,steps=1000)
```

## Evaluation:

### model.evaluate()
`.evaluate()` method is takes an `input function` & `number of steps` as its function parameters. This method evaluates various parameters such as `accuracy`, `loss`, `precision`,etc.

```python
results=model.evaluate(input_fn=input_train_func)
```

```python
results
```

## Predictions:

### model.predict()
`.predict()` method takes an `input function` as its function parameters and generates predictions for given `output function`.

```python
predictions=model.predict(input_fn=input_test_func)
```

```python
list(predictions)
```
